# London Housing and Population Analysis (2002 - 2015 data)

## 7. Pops, Sales, Prices Correlation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/csmyth/Desktop/Code/london_housing_eda/pops_sales_and_prices.csv', index_col=[0])

# Ascertain correlations between Year, Population, House Sales, and Sell Prices:

grid = sns.pairplot(df, vars=['year', 'all_ages', 'houses_sold', 'average_price'], kind="reg", diag_kind=None, 
plot_kws={'color': 'dimgrey', 'line_kws':{'color':'red'}}, diag_kws={'color': 'dimgrey'})

years = df['year'].unique().tolist()
grid.axes[0, 0].set_xticks([years[3], years[8], years[13]])
grid.axes[0, 3].set_yticks([years[3], years[8], years[13], 2020])

grid.axes[1, 0].set_facecolor('#ffffd4')
grid.axes[2, 0].set_facecolor('#ffffd4')
grid.axes[3, 0].set_facecolor('#ffffd4')

plt.subplots_adjust(top=0.95, left=0.125)
grid = grid.fig.suptitle('Figure 7')
plt.savefig('Fig7_pops_sales_prices_pairplot.png')
plt.show()
```

```python
# Create Fig7b to show only relationships between year and average price, houses sold, and population (one figure each)
figure, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(16, 4), gridspec_kw={'wspace': 0.5})

ax1.set_facecolor('#ffffd4')
ax2.set_facecolor('#ffffd4')
ax3.set_facecolor('#ffffd4')

sns.regplot('year', 'average_price', data=df, color='dimgrey', line_kws={'color':'red'}, ax=ax1)
sns.regplot('year', 'houses_sold', data=df, color='dimgrey', line_kws={'color':'red'}, ax=ax2)
sns.regplot('year', 'all_ages', data=df, color='dimgrey', line_kws={'color':'red'}, ax=ax3)

plt.subplots_adjust(top=0.9, left=0.125)
plt.suptitle('Figure 7b\n')

plt.savefig('Fig7b_year_sales_prices_relationship.png')
plt.show()
```

_Intuitive correlations/trends:_  
1. As house sales increase, average house price decreases;  
2. As population increases, house sales increase.  
   Both only slight trends (low gradients of regression lines).  

_Other observations:_  
a) Two separate increases in house sales - 2002-2007 / 2008-2015.  -   Sales activity hadn't returned to pre-2008 level by 2015;   Sell price had exceeded 2008 level in some boroughs;   
b) Increasing number of expensive price outliers in recent years -  Increasing gap in purchase price between majority of boroughs and most expensive boroughs.
